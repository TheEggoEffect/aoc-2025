```python
#Kenton's method
#!/bin/bash
def main(dial=50, m=100, path="aoc2025 day 1 input.txt"):
    zero_count = 0
    with open(path) as fd:
        for line in fd:
            amt = int(line[1:])

            start_pos = dial
            q, dial = divmod(dial-amt if line[0] == "L" else dial+amt, m)
            zero_count += abs(q)
            
            # fix annoying fucking rotate left behavior
            if line[0] == "L":
                if start_pos == 0 and dial != 0:
                    zero_count -= 1  # started at 0, don't count
                elif start_pos != 0 and dial == 0:
                    zero_count += 1  # landed on 0, do count
            
    print(zero_count)

if __name__ == "__main__":
    main()
```

```python
#My method for part 2:
file = open("aoc2025 day 2 input.txt")
a = file.read()
file.close()
a = a.replace("-", ",")
a = a.split(",")
#Opens text file and reads it as a string, then takes the string apart into individual number strings for manipulation

ids = []
for i in range(len(a) // 2):
    lo = int(a[2 * i])
    hi = int(a[(2 * i) + 1])
    inc = lo
    while inc != hi + 1:
        ids.append(str(inc))
        inc += 1
#Takes each pair of IDs (representing upper and lower bounds) and appends every integer between them into a master list of IDs

def factors(n):
    f = []
    for i in range(1, n // 2 + 1):
        if n % i == 0:
            f.append(i)
    return f
#Returns a list containing the factors of an integer, up to half the integer (since the below doesn't work for f > n/2)

inval = 0
for i in range(len(ids)):
    f = factors(len(ids[i]))
    for j in range(len(f)):
        u = ids[i][:f[j]].split(" ")[0]
        q = len(ids[i]) // f[j]
        w = u * q
        if w == ids[i]:
            inval += int(ids[i])
            break
#Runs through the list of IDs, calculating the factors of each ID's length. Takes all series of numbers whose lengths are one of these factors (as these
#are the chunks that can be copied to fill the entire ID). Then, check if an ID filled with copied chunks is the same as the ID itself, and adds this to
#the total. 
print(inval)
```

```python
#My method for part 2:
file = open("aoc2025 day 3 input.txt")
a = file.read()
file.close()
a = a.split("\n")
#Turns the input file into a list of big ol strings

totaljolt = 0
for i in range(len(a)):
    b = a[i]
    digits = []
    for j in range(12, 0, -1):
        c = []
        for k in range(len(b) - j + 1):
            c.append(int(b[k]))
        digits.append(max(c))
        ind = c.index(max(c))
        b = b[ind + 1:]
    for j in range(1, len(digits) + 1):
        totaljolt += digits[-j] * 10**(j - 1)
#Goes down each bank of batteries, selecting the highest joltage that can serve as the 10^11 digit. Stores this joltage and slices off all digits before
#it, then repeats the process to find the 10^10 digit, and so on until all digits are accounted for. Adds the number formed from this list of digits to
#the total joltage.

print(totaljolt)
```

```python
#My method for part 1:

#You know what this does
file = open("aoc2025 day 6 input.txt")
inp = file.read()
file.close()
inp = inp.split("\n")
for i in range(len(inp)):
    inp[i] = inp[i].split(" ")
    while "" in inp[i]:
        inp[i].remove("")

#Rearranges each line of the problem so everything is in the same list
problems = []
for i in range(len(inp[0])):
    line = []
    for j in inp:
        line.append(j[i])
    problems.append(line)

#My attempts to convert all strings representing ints into ints back in the beginning failed, so here's a custom sum(list) function that works with strings. (I would've had to make one for product(list) anyways)
def sum(x):
    s = 0
    for i in x:
        s += int(i)
    return s
def product(x):
    p = 1
    for i in x:
        p *= int(i)
    return p

#Goes through each problem, adding either the sum or product of every number in the problem to a running total, depending on if a + or * is indicated
total = 0
for i in problems:
    if i[-1] == "+":
        total += sum(i[:-1])
    if i[-1] == "*":
        total += product(i[:-1])

print(total)
```
